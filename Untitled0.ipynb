Importing the libraries and loading iris dataset

```python
#Importing the requuired libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz 
from sklearn.model_selection import train_test_split
import sklearn.metrics as sm

import pandas as pd 
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import pydot 
from IPython.display import Image

import warnings
warnings.filterwarnings('ignore')
```

Loading the Dataset

```python
#Loading dataset 
iris = load_iris()
x=iris.data[:,:]
y=iris.target
```

Exploratory Data Analysis

```python
#Input data

data=pd.DataFrame(iris['data'],columns=["Petal lengeth","Petal Width","Sepal Length","Sepal width"])
data['Species']=iris['target']
data['Species']=data['Species'].apply(lambda x:iris['target_names'][x])

data.head()
```

```python
data.shape
```

```python
data.describe()
```

Data Visualization comparing various features

```python
#Input data Visualization
sns.pairplot(data)
```

```python
#Scatter plot of data based on Sepal Length and width features
sns.FacetGrid(data,hue='Species').map(plt.scatter,'Sepal Length','Sepal width').add_legend()
plt.show()
```

Decision Tree Model Training

```python
#Model Training
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.33,random_state=42)
tree_classifier =DecisionTreeClassifier()
tree_classifier.fit(x_train,y_train)
print("Training Complete.")
y_pred = tree_classifier.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

Visualizing the Trained Model

```python
#visualizing the trained Decision Tree Classifier taking all 4 features in consideration

export_graphviz(
    tree_classifier,
    out_file="img\decision_tree.dot",
    feature_names=iris.feature_names[:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
(graph,) = pydot.graph_from_dot_file('img\decision_tree.dot')
graph.write_png('img\decision_tree.png')

Image(filename='img\decision_tree.png')
```

Import the Decision Tree Classifier and fit the model

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)

print('Decision Tree Classifier Created Successfully')
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

a=['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)']
b=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)

tree.plot_tree(dtree,feature_names=a, class_names=b, filled=True);
```

Predicting the class output for some random values of petal and sepal length and width

```python
print("Class Names =",iris.target_names)

#Estimating class probabilities
print()
print("Estimating Class Probabilities for flower whose petals length width are 4.7cm and 3.2cm and sepal length and width are 1.3cm and 0.2cm. ")
print()
print('output =',tree_classifier.predict([[4.7, 3.2, 1.3, 0.2]]))
print()
print("Ou model predicts theclss as 0, that is, setosa.")

```

Calculating The Model accuracy

```python
print("Accuracy:",sm.accuracy_score(y_test,y_pred))
```
